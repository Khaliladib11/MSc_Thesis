# Import Libraries

```python
import albumentations as A
from PIL import Image
import os
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_to_images = '../../dataset/bdd100k/images/10k/bdd100k/images/10k/train'
path_to_masks = '../../dataset/bdd100k/labels/ins_seg/colormaps/train'
```

```python
images = os.listdir(path_to_images)
masks = os.listdir(path_to_masks)
```

```python
idx = random.randint(0, len(images))
img = Image.open(os.path.join(path_to_images, images[idx]))
mask = Image.open(os.path.join(path_to_masks, images[idx].replace('.jpg', '.png')))

fix, ax = plt.subplots(1, 2, figsize=(15,15))
ax[0].imshow(img)
ax[1].imshow(mask)
```

```python
transform = A.Compose([
    A.HorizontalFlip(p=1),
])

transformed = transform(image=np.array(img), mask=np.array(mask))

transformed_img = transformed['image']
transformed_mask = transformed['mask']
fix, ax = plt.subplots(1, 2, figsize=(15,15))
ax[0].imshow(transformed_img)
ax[1].imshow(transformed_mask)
```

```python
transform = A.Compose([
    A.Flip(),
])

transformed = transform(image=np.array(img), mask=np.array(mask))

transformed_img = transformed['image']
transformed_mask = transformed['mask']
fix, ax = plt.subplots(1, 2, figsize=(15,15))
ax[0].imshow(transformed_img)
ax[1].imshow(transformed_mask)
```

```python
transform = A.Compose([
    A.CropAndPad(px=100, p=0.5),
])

transformed = transform(image=np.array(img), mask=np.array(mask))

transformed_img = transformed['image']
transformed_mask = transformed['mask']
fix, ax = plt.subplots(1, 2, figsize=(15,15))
ax[0].imshow(transformed_img)
ax[1].imshow(transformed_mask)
```

```python
transform = A.Compose([
    A.RandomCrop(500, 500),
])

transformed = transform(image=np.array(img), mask=np.array(mask))

transformed_img = transformed['image']
transformed_mask = transformed['mask']
fix, ax = plt.subplots(1, 2, figsize=(15,15))
ax[0].imshow(transformed_img)
ax[1].imshow(transformed_mask)
```
