# Import Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from config.defaults import cfg
from collections import deque
import json
import random
from dataset.bdd_instance_segmentation import BDDInstanceSegmentation
import matplotlib.patches as patches
```

# Load annotations

```python
%%time
bdd_train_params = {
    'cfg': cfg,
    'stage': 'train'
}
bdd_train = BDDInstanceSegmentation(**bdd_train_params)
```

```python
len(bdd_train)
```

# Test

```python
idx = random.randint(0, len(bdd_train))
image = bdd_train.get_image(idx)
mask = bdd_train.get_mask(idx)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 15))
ax1.imshow(image)
ax2.imshow(mask)
```

# Test Get Masks

```python
idx = random.randint(0, len(bdd_train))
targets = bdd_train._get_labels(idx)
```

```python
image = Image.open(bdd_train.db[idx]['image_path'])
for i in range(0, len(targets['labels'])):
    mask = targets['masks'][i]
    label = targets['labels'][i]
    print(bdd_train.idx_to_cls[label])
    box = targets['boxes'][i]
    fig, ax = plt.subplots(1, 2, figsize=(15, 15))
    ax[0].imshow(image)
    rect = patches.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1],
                                         edgecolor='red',
                                         facecolor="none", linewidth=1)
    ax[0].add_patch(rect)

    ax[1].imshow(mask)
    ax[0].axis('off')
    ax[1].axis('off')
    plt.show()
```

# Test __getitem__ 

```python
idx = random.randint(0, len(bdd_train))
original_image = bdd_train.get_image(idx, False)
image, target = bdd_train[idx]
```

```python
plt.imshow(image.permute(1, 2, 0))
```

```python
for i in range(0, len(target['boxes'])):
    mask = target['masks'][i]
    label = target['labels'][i]
    print(label)
    box = target['boxes'][i]
    fig, ax = plt.subplots(1, 3, figsize=(15, 15))
    ax[0].imshow(original_image)
    ax[1].imshow(image.permute(1, 2, 0))
    rect = patches.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1],
                                         edgecolor='red',
                                         facecolor="none", linewidth=1)
    ax[1].add_patch(rect)

    ax[2].imshow(mask)
    ax[0].axis('off')
    ax[1].axis('off')
    ax[2].axis('off')
    plt.show()
```
