```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import shutil
```

```python
def plot_predictions(demo_folder_name):
    demo_path = '../weights/demo'
    path = os.path.join(demo_path, demo_folder_name)
    images = os.listdir(path)
    for image in images:
        plt.figure(figsize=(10, 10))
        img = Image.open(os.path.join(path, image))
        plt.imshow(img)
        plt.title(demo_folder_name + image.replace('.jpg', ''))
        plt.axis('off')
        plt.show()
```

```python
plot_predictions('example1')
```

```python
plot_predictions('example6')
```

```python
plot_predictions('example18')
```
